### MISSING VALUE TREATMENT

```python
## treating the missing values using mode or mean or knn
##if numerical dta we fill it with median or mean or knn imputer and if it categorical we fill it with mode
data.isnull().sum()
```

```python
for i in ['BMI','Polio']:
    data[i].fillna(data[i].median(),inplace=True)
 
```

```python
data.isnull().sum()
```

```python
data.select_dtypes(include='number').columns
```

```python
## missing value treatment using knn imputer
data.columns = data.columns.str.strip()
impute=KNNImputer()
for i in ['Year','Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles','BMI','under-five deaths','Polio','Total expenditure','Diphtheria','HIV/AIDS','GDP','Population','thinness  1-19 years','thinness 5-9 years','Income composition of resources','Schooling']:
    data[i]=impute.fit_transform(data[[i]]).ravel()
    
```

### OUTLIER TREATMENT

```python
data.columns
```

```python
def wrisker(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lw = Q1 - 1.5 * IQR   # lower whisker
    uw = Q3 + 1.5 * IQR   # upper whisker
    return lw, uw
```

```python
data.columns = data.columns.str.strip()

for i in ['BMI','Total expenditure','thinness  1-19 years','thinness 5-9 years']:
    lw, uw = wrisker(data[i])
    data[i] = np.where(data[i] < lw, lw, data[i])   # cap lower outliers
    data[i] = np.where(data[i] > uw, uw, data[i])   # cap upper outliers
```

```python
for i in ['BMI','Total expenditure','thinness  1-19 years','thinness 5-9 years']:
    sns.boxplot(data[i])
    plt.show()
```
